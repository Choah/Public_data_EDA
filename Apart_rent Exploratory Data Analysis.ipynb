```python
'''

How to process observations without values?

1. Remove the observations with Null value
2. Delete the entire feature
3. Fill with some value(0,mean,median ...)

--> Use option 1,3 fitting to this data

경상북도 칠곡군 석적읍 남율로9길 61(칠곡남율효성해링턴플레이스2단지)
경상북도 칠곡군 석적읍 남율리 1337 칠곡남율효성해링턴플레이스2단지

경상북도 칠곡군 석적읍 남율로9길 81(칠곡남율효성해링턴플레이스3단지)
경상북도 칠곡군 석적읍 남율리 1511 칠곡남율효성해링턴플레이스3단지

세종특별자치시 마음안로 34(고운동, 가락마을5단지)
세종특별자치시 고운동 1705 가락마을5단지

경기도 의정부시 오목로 251(낙양동, 부용마을 3단지)
경기도 의정부시 낙양동 712 부용마을 3단지

Fill NA values in column ['번지'] using option 3

세종특별자치시 아름동 1-2생활권, 1-4생활권은 검색 결과 나오지 않음 --> Remove obs using option 1

'''

# option3를 실행한다[단 번지수가 검색이 되는 경우] #

rent15_jeonse.reset_index()
rent15_jeonse.loc[rent15_jeonse['번지'].isnull(),'번지'] = np.array([
    np.nan,np.nan,np.nan,
    1511,1511,1511,
    1511,1705,1511,
    1511,1511,1511,
    1511,1511,1511,
    1511,1511,712])

rent15_jeonse[rent15_jeonse['번지'].isnull()]
```
